  * Kelas: 4 Solanum tuberosum, 4 Solanum melongena

### 4. Hitung Entropy Awal

$$
p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}
$$

$$
Entropy(D) = -\frac{8}{15}\log_2\frac{8}{15} - \frac{7}{15}\log_2\frac{7}{15} \approx 0.996
$$

---

### 5. Entropy Setelah Split

#### Group Left (≤ 16 cm):

* 4 tuberosum, 4 melongena → balanced

$$
Entropy(left) = -\frac{4}{8}\log_2\frac{4}{8} - \frac{4}{8}\log_2\frac{4}{8} \\
= -2 \cdot \frac{4}{8} \log_2 \frac{4}{8} = -2 \cdot 0.5 \cdot (-1) = \boxed{1.0}
$$

#### Group Right (> 16 cm):

* 4 tuberosum, 3 melongena

$$
Entropy(right) = -\frac{4}{7}\log_2\frac{4}{7} - \frac{3}{7}\log_2\frac{3}{7} \\
≈ -0.571 \cdot (-0.807) - 0.429 \cdot (-1.222) \\
≈ 0.461 + 0.524 = \boxed{0.985}
$$

---

### 6. Information Gain

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot 1.0 + \frac{7}{15} \cdot 0.985 \right) \\
= 0.996 - (0.533 + 0.459) \\
= 0.996 - 0.992 = \boxed{0.004}
$$

---

### 7. Gini Index

#### Gini Awal:

$$
Gini(D) = 1 - \left( \frac{8}{15} \right)^2 - \left( \frac{7}{15} \right)^2 = 1 - 0.284 - 0.218 = \boxed{0.497}
$$

#### Gini Left:

* 4 vs 4

$$
Gini(left) = 1 - \left( \frac{4}{8} \right)^2 - \left( \frac{4}{8} \right)^2 = 1 - 0.25 - 0.25 = \boxed{0.5}
$$

#### Gini Right:

* 4 vs 3

$$
Gini(right) = 1 - \left( \frac{4}{7} \right)^2 - \left( \frac{3}{7} \right)^2 \\
= 1 - 0.327 - 0.184 ≈ \boxed{0.489}
$$

#### Gini Split:

$$
Gini_{split} = \frac{8}{15} \cdot 0.5 + \frac{7}{15} \cdot 0.489 ≈ 0.267 + 0.228 = \boxed{0.495}
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Sayur ===
data = {
    "Panjang Sayur (cm)": [21, 15, 20, 13, 12, 12, 14, 20, 17, 13, 22, 19, 14, 16, 19],
    "Berat Sayur (g)":    [210, 140, 200, 100, 120, 133, 187, 190, 205, 160, 220, 140, 159, 135, 200],
    "Jenis": [
        "Solanum tuberosum", "Solanum tuberosum", "Solanum tuberosum", "Solanum tuberosum", "Solanum tuberosum",
        "Solanum melongena", "Solanum melongena", "Solanum melongena", "Solanum melongena", "Solanum melongena",
        "Solanum tuberosum", "Solanum tuberosum", "Solanum melongena", "Solanum tuberosum", "Solanum melongena"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Jenis'])  # tuberosum=1, melongena=0 (urutan bisa berbeda tergantung encode)

# === Fitur dan Target ===
X = df[["Panjang Sayur (cm)", "Berat Sayur (g)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## Visualisasi Splitting

```python
# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Panjang Sayur (cm)': [14, 19],
    'Berat Sayur (g)': [150, 210]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (p, b) in enumerate(zip(uji_baru['Panjang Sayur (cm)'], uji_baru['Berat Sayur (g)'])):
    print(f"Data {i+1}: Panjang={p}, Berat={b} → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

x_min, x_max = X['Panjang Sayur (cm)'].min() - 2, X['Panjang Sayur (cm)'].max() + 2
y_min, y_max = X['Berat Sayur (g)'].min() - 10, X['Berat Sayur (g)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Sayur (cm)', 'Berat Sayur (g)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Sayur (cm)'], X.iloc[idx]['Berat Sayur (g)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.scatter(uji_baru['Panjang Sayur (cm)'], uji_baru['Berat Sayur (g)'],
            c='black', marker='X', s=100, label='Data Uji')

plt.xlabel("Panjang Sayur (cm)")
plt.ylabel("Berat Sayur (g)")
plt.title("Decision Boundary (Panjang vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

x_min, x_max = X['Panjang Sayur (cm)'].min() - 10, X['Panjang Sayur (cm)'].max() + 10
y_min, y_max = X['Berat Sayur (g)'].min() - 10, X['Berat Sayur (g)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Sayur (cm)', 'Berat Sayur (g)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Sayur (cm)'], X.iloc[idx]['Berat Sayur (g)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji
plt.scatter(uji_baru['Panjang Sayur (cm)'], uji_baru['Berat Sayur (g)'],
            c='black', marker='X', s=100, label='Data Uji')

plt.xlabel("Panjang Sayur (cm)")
plt.ylabel("Berat Sayur (g)")
plt.title("Decision Boundary (Panjang vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
